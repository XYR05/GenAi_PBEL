```python
### Basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
```

```python
def collect_data():
    images=[]
    labels=[]
    label_dic={}
    people=[person for person in os.listdir("DataSet1")]
    for i,person in enumerate(people):
        label_dic[i]=person
        for img in os.listdir("DataSet1/"+person):
            if img.endswith(".jpg"):
                imgs=cv2.imread("DataSet1/"+person+"/"+img,0)
                images.append(imgs)
                labels.append(i)
    return (images,labels,label_dic)
```

```python
images,target,label_dic=collect_data()
```

```python
print(len(images))
print(len(target))
```

```python
label_dic
```

```python
pd.Series(target).value_counts()
```

```python
x=np.array(images)
y=np.array(target)
```

```python
x_train=x.reshape(len(x),-1)
x_train
```

### Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

- (x-u)/std

```python
scaled_data=StandardScaler()
data=scaled_data.fit_transform(x_train)
```

```python
pca=PCA(0.98)
final_data=pca.fit_transform(data)
```

```python
pca.n_components_
```

```python
### SVM
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, cross_val_score,KFold
```

```python
scv=SVC(probability=True)
```

```python
para={
    "C":[0.0001,0.001,0.01,0.1,1],
    "kernel":['rbf',"linear"]
}
```

```python
kf=KFold(5,shuffle=True)
```

```python
final_model=GridSearchCV(scv,param_grid=para,cv=kf,scoring="accuracy")
```

```python
final_model.fit(final_data,y)
```

```python
print(final_model.best_params_)
print(final_model.best_score_)
```

```python
fsm=final_model.best_estimator_
```

```python
fsm.fit(final_data,y)
```

```python
import pickle
```

```python
file1="face_recog_model.pkl"
f1=open(file1,'wb')
pickle.dump(fsm,f1)
f1.close()

print("Face_recog model saved !!!")
```

```python
file2="scaling_model.pkl"
f2=open(file2,"wb")
pickle.dump(scaled_data,f2)
f2.close()
print("Scaling model saved !!!")
```
