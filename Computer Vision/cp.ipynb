### Gray scale

```python
def gray_scale(image):
    cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Cut Face

```python
def cut_face(image, face_coord):
    cut_faces= []
    for (x, y, w, h) in face_coord:
        face= image[y:y+h, x:x+w]
        cut_faces.append(face)
    return cut_faces
```

### Resize

```python
def resize(images, size=(80,100)):
    resized_images= []
    for image in images:
        img= cv2.resize(image, size)
        resized_images.append(img)
    return resized_images
```

### Normalize Intensity

```python
def normalize_intensity(images):
    normalized_faces= []
    for image in images:

        normalized_faces.append(cv2.equalizeHist(image))
    return normalized_faces
```

# IMAGE PLOT

```python
def plot_images(image, title=''):
    plt.figure(figsize=(10, 10))
    if image.shape==3:
        cv2.cvtColor(image,cv2.COLOR_BGR2BGR555)
    plt.imshow(image, c='gray')
    plt.title(title)
    plt.axes("off")
    plt.show()
```

# Pipeline

```python
def pipeline(image,face_coord):
    faces=cut_face(image,face_coord)
    faces=resize (faces)
    faces=normalize_intensity(faces)
    return faces
```
