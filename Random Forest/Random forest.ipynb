# feature randomness 
is a technique used in machine learning to improve the performance of models by introducing randomness into the feature selection process. This can help to reduce overfitting and improve generalization by ensuring that the model does not rely too heavily on any one feature or subset of features.

# bagging 
is a ensemble technique which is used to reduce the variance of the model by combining the predictions of multiple models. It is also known as bootstrap aggregating. In bagging, we create multiple subsets of the original dataset with replacement and train a model on each subset. The final prediction is made by averaging the predictions of all the models.

# boosting
it is a machine learning technique that is used to improve the accuracy of a model by combining multiple weak models to create a strong model. The idea behind boosting is to train a series of weak models, each of which focuses on the mistakes made by the previous models, and then combine their predictions to make a final prediction.

# stacking
it is a technique in machine learning where multiple models are combined to improve the overall performance of the prediction. The idea is to train several base models on the same dataset and then use their predictions as input for a final model, which makes the final prediction. This approach can help to reduce overfitting and improve generalization by leveraging the strengths of different models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.ensemble import RandomForestClassifier
x=[[2],[3],[4],[5],[6],[7],[8]]
y=["fail","fail","fail","pass","pass","pass","pass"]
model=RandomForestClassifier(n_estimators=10, random_state=50)
model.fit(x,y)
print(model.predict([[5]]))
```

```python
x=[[120],[130],[140],[180],[190],[200]]
y=["apple","apple","apple","orange","orange","orange"]
model=RandomForestClassifier(n_estimators=10, random_state=50)
model.fit(x,y)
print(model.predict([[150]]))
```
