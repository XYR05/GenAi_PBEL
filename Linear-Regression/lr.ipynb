```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/home/harry/Internship - ibm/Linear-Regression/Position Salary.csv')
df
```

```python
x=df[['Level']]
y=df['Salary']  
```

Model Building


```python
model=LinearRegression()
model.fit(x, y)  ## fit is use to calc the parameters of the model
pred=model.predict(x)
pred
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, pred, color='blue')
plt.title('Level vs Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
x_poly=PolynomialFeatures(degree=4) #this is a method used to create polynomial features from the original input data. The degree parameter specifies the degree of the polynomial features to be generated. In this case, degree=2 means that the method will generate polynomial features up to the second degree (i.e., x^2). This allows the model to capture non-linear relationships in the data.
x_poly=x_poly.fit_transform(x) #this method is used to transform the input data into polynomial features. It takes the original input data (x) and generates new features that are polynomial combinations of the original features. For example, if the original feature is x, the transformed features could include x^2, x^3, etc., depending on the degree specified when creating the PolynomialFeatures object. This transformation allows linear models to fit non-linear relationships in the data.
x_poly
```

```python
lr=LinearRegression()
lr.fit(x_poly, y)
pr=lr.predict(x_poly)
pr
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, pr, color='blue')
plt.title('Level vs Salary')
plt.show()
```
