```python
import pandas as pd
import numpy as np
#Read csv file
df = pd.read_csv("/home/harry/Internship - ibm/Assignment-3/temperature.csv")
df.columns = df.columns.str.strip() # Clean column names
df
```

```python
#using numpy to calculate highest,lowest and average temperature
highest_temp = np.max(df["Temperature"])
lowest_temp = np.min(df["Temperature"])
avg_temp = np.mean(df["Temperature"])

print("Highest temperature:", highest_temp)
print("Lowest temperature:", lowest_temp)
print("Average temperature:", avg_temp)
```

```python
#function to classify temperature
def classify_temp(temp):
    if temp < 20:
       return "Cold"
    elif temp <= 30:
         return "Moderate"
    else:
         return "Hot"
for temp in df["Temperature"]:
  print(temp, classify_temp(temp))
```

```python
# Display category for each day
print("\nCategory for each day:")
for index, row in df.iterrows():
    print(row["Day"], "-", classify_temp(row["Temperature"]))
```
