```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.neighbors import KNeighborsClassifier
x=np.array([[1],[2],[3],[6],[7],[8]])
y=np.array(["Fail","Fail","Fail","Pass","Pass","Pass"])
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x,y) #why in knn we use fit method?
# In K-Nearest Neighbors (KNN), the `fit` method is used to train the model on the provided training data. When you call `model.fit(x, y)`, you are essentially telling the KNN algorithm to store the training data (features `x` and labels `y`) for later use in making predictions.

prediction=model.predict([[3.5]])
print(prediction)

```

# why fail is output for 3.5 in knn?
In K-Nearest Neighbors (KNN), the prediction for a new data point is determined by looking at the 'k' nearest neighbors in the training data and taking a majority vote among their labels. In this case, you have set `n_neighbors=3`, which means the algorithm will consider the three closest points to the input value of 3.5.

```python
x=[[1,2],[2,3],[3,1],[6,7],[7,8],[8,6]]
y=[0,0,0,1,1,1]
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
print(knn.predict([[5,5]])) 

```

# why 1 is output of above code?
The output of `1` from the above code is due to how the K-Nearest Neighbors (KNN) algorithm works. In this case, the model is trained on a dataset with two classes: `0` and `1`. The training data consists of points that belong to these two classes. When you call `knn.predict([[5,5]])`, the algorithm looks for the  three nearest neighbors to the point `[5,5]` in the training data. The three closest  points ([6,7],[7,8],[8,6]) are closest to `[5,5]` are likely to be from the class `1`, which is why the prediction returns `1`.

```python
x=[[1],[2],[3],[4],[5],[6]]
y=["fail","fail","fail","pass","pass","pass"]
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
print(knn.predict([[3.5]]))

```

# why fail is predicted for 3.5?

In K-Nearest Neighbors (KNN), the prediction for a new data point is based on the majority class of its k-nearest neighbors in the training data. In this case, when you predict for the value 3.5, the algorithm looks at the three nearest neighbors in the training data, which are 3, 4, and 5. Among these neighbors, two of them (3)  belong to the "fail" class, while  (4, 5) belongs to the "pass" class.  now we will check for even or odd vlaue of nearest index if majority is even then we will take the nearest index value and if it is odd then we will take the majority class value. In this case, the nearest index value is 3 which is odd so we will take the majority class value which is "fail".
    

```python
from sklearn.neighbors import KNeighborsRegressor
x=[[1],[2],[3],[4],[5]]
y=[10,20,30,40,50]
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x,y)
print(knn.predict([[3.5]]))

```

# why output is 30.0 in knn regression? 
the output is 30.0 in KNN regression because the model predicts the value for the input [[3.5]] by averaging the values of its 3 nearest neighbors  in the training data.

# what is the difference between KNeighborsClassifier and KNeighborsRegressor?
The main difference between `KNeighborsClassifier` and `KNeighborsRegressor` lies in the type of problem they are designed to solve:
1. **KNeighborsClassifier**: This is used for classification problems, where the goal is to predict a categorical label (class) for a given input based on the majority class of its nearest neighbors. It outputs discrete class labels (e.g., "pass" or "fail").
2. **KNeighborsRegressor**: This is used for regression problems, where the goal is to predict a continuous numerical value for a given input based on the average (or weighted average) of its nearest neighbors. It outputs continuous values (e.g., predicting a price or temperature).
