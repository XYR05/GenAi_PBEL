```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =sns.load_dataset('iris')
data
```

Label encoding and one hot encoding can be used.

```python
from sklearn.datasets import load_iris
data1= load_iris()
data1
```

```python
x=data1['data']
y=data1['target']
# converted 2d data into 1d data beacause we have to split the data into train and test data
```

```python
x
```

```python
y
```

### Data Splition part

```python
from sklearn.model_selection import train_test_split
```

if need validation divide data into 3 parts test , train , validiate

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

# Model building part

```python
from sklearn.tree import DecisionTreeClassifier  
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train) ## training the model : why we are using x_train and y_train because we have to train the model on the training data and then we will test the model on the test data 
```

```python
# checking the how much it have learned from the training data and how much it is able to predict the test data by plotting the decision tree
from sklearn import tree 
plt.figure(figsize=(15,10)) ## setting the outline and figure size
tree.plot_tree(model, filled=True) ## plotting the decision tree
```

# need of pruning :
now since the output was clear on condition x[3]<=1.6 and depth of the tree was 3 so we will limit the depth of the tree to 2 and see how it performs

# Making predictions without pruning.

```python
y_pred=model.predict(x_test) ## predicting the test data
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("test accuracy:",accuracy_score(y_test,y_pred))
print("confusion matrix:",confusion_matrix(y_test,y_pred))
```

## Pruning :
 **1  Pre Pruning:**
    now we want to stop decision tree from overfitting the data so we will use the max_depth parameter to limit the depth of the tree and we will also use the min_samples_split parameter to limit the number of samples required to split an internal node and we will also use the min_samples_leaf parameter to limit the number of samples required to be at a leaf node

```python
model1=DecisionTreeClassifier(max_depth=2)
model1.fit(x_train,y_train)
tree.plot_tree(model1, filled=True) ## plotting the decision tree
```

```python
y_pred1=model1.predict(x_test) ## predicting the test data
print("test accuracy:",accuracy_score(y_test,y_pred1))
print("confusion matrix:",confusion_matrix(y_test,y_pred1))
```
